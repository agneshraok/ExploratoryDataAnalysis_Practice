# Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing the data and Checking it out

```python
top_50 = pd.read_csv('spotify_top50_2021.csv')
top_50.head()
```

```python
top_50
```

```python
top_50.shape
```

# Dealing with Missing Data


```python
top_50['artist_name'][2]
```

```python
top_50[top_50.isna().any(axis=1)]  # Using isna() to select all rows with NaN under an entire DataFrame
```

```python
cleaned1_top50 = top_50.dropna()
cleaned1_top50.shape
```

```python
# the feature duration_ms has to be converted to mins

cleaned1_top50['duration_ms'] = cleaned1_top50['duration_ms'].div(60000)
cleaned1_top50
```

```python
cleaned2_top50 = cleaned1_top50.rename(columns={"duration_ms":"duration_mins"})
cleaned2_top50
```

# Visualizations : Histogram

```python
%matplotlib inline
```

```python
plt.hist(cleaned2_top50.popularity,bins = 100)
```

# Visualizations : Subplots

```python
plt.figure(figsize=(20,20)) 

plt.subplot(5,2,1)
plt.plot(cleaned2_top50.danceability,cleaned2_top50.popularity, ls ='', marker ='o')
plt.xlabel('danceability')
plt.ylabel('popularity')
plt.title('popularity vs danceability')
plt.grid(True)


plt.subplot(5,2,2)
plt.plot(cleaned2_top50.energy,cleaned2_top50.popularity, ls ='', marker ='o')
plt.xlabel('energy')
plt.ylabel('popularity')
plt.title('popularity vs energy')
plt.grid(True)

plt.subplot(5,2,3)
plt.plot(cleaned1_top50.loudness,cleaned2_top50.popularity, ls ='', marker ='o')
plt.xlabel('loudness')
plt.ylabel('popularity')
plt.title('popularity vs loudness')
plt.grid(True)

plt.subplot(5,2,4)
plt.plot(cleaned2_top50.speechiness,cleaned2_top50.popularity, ls ='', marker ='o')
plt.xlabel('speechiness')
plt.ylabel('popularity')
plt.title('popularity vs speechiness')
plt.grid(True)

plt.subplot(5,2,5)
plt.plot(cleaned2_top50.acousticness,cleaned2_top50.popularity, ls ='', marker ='o')
plt.xlabel('acousticness')
plt.ylabel('popularity')
plt.title('popularity vs acousticness')
plt.grid(True)

plt.subplot(5,2,6)
plt.plot(cleaned2_top50.instrumentalness,cleaned2_top50.popularity, ls ='', marker ='o')
plt.xlabel('instrumentalness')
plt.ylabel('popularity')
plt.title('popularity vs instrumentalness')
plt.grid(True)


plt.subplot(5,2,7)
plt.plot(cleaned2_top50.liveness,cleaned2_top50.popularity, ls ='', marker ='o')
plt.xlabel('liveness')
plt.ylabel('popularity')
plt.title('popularity vs liveness')
plt.grid(True)


plt.subplot(5,2,8)
plt.plot(cleaned2_top50.tempo,cleaned2_top50.popularity, ls ='', marker ='o')
plt.xlabel('tempo')
plt.ylabel('popularity')
plt.title('popularity vs tempo')
plt.grid(True)

plt.subplot(5,2,9)
plt.plot(cleaned2_top50.duration_mins,cleaned2_top50.popularity, ls ='', marker ='o')
plt.xlabel('duration_mins')
plt.ylabel('popularity')
plt.title('popularity vs duration_mins')
plt.grid(True)


#to set the gap between subplots
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
```

# Significance Testing

```python
population_mean = cleaned2_top50.speechiness.mean()
population_mean
```

```python
standard_deviation = cleaned2_top50.speechiness.std()
standard_deviation
```

```python
sample = cleaned2_top50[cleaned2_top50.speechiness < 0.1]
sample.head()
```

```python
sample.shape
```

```python
sample_mean = sample.speechiness.mean()
sample_mean
```

```python
z_score = (population_mean - sample_mean)/(standard_deviation/50)
z_score
```
